## Relationship Analysis

### Numeric Vs Numeric Data 

```python
correlation = df.corr()
sns.heatmap(correlation, xticklabels = correlation.columns, yticklabels = correlation.columns, annot=True)
```

```python
sns.pairplot(df)
```

### Numeric Vs Numeric vs Categorical 

```python
sns.relplot(x= 'wealth_segment', y = 'tenure', hue = 'gender', data = df)
```

```python
sns.distplot(df['tenure'], bins = 5)
```

### Numeric Vs Categorical 

```python
sns.catplot(x='wealth_segment', y='tenure',kind = 'box', data= df)
```
