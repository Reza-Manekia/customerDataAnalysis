# DATA CLEANING AND PREPROCESSING

```python
import pandas as pd
import numpy as np


data = pd.read_csv("D:\CustomerData.csv")
data.head()
```

```python
Data = data.drop(['Unnamed: 13','default'], axis=1)
Data
```

```python
Data.nunique()
```

```python
Data['gender'].unique()
```

```python
# In gender there are 6 categories we can split into F (for females) and M (for males) 
# and drop having gender "U" by using following code:
```

```python
gender_map = {'Female': 'F', 'Male': 'M', 'Femal': 'F'}
Data['gender'] = Data['gender'].replace(gender_map)

indices_to_drop = []

for index, row in Data.iterrows():
    if row['gender'] == 'U':
        indices_to_drop.append(index)

Data = Data.drop(indices_to_drop)
Data
```

```python
Data['gender'].unique()
```

```python
# Tenure and DOB have approximately 87 null values we can drop from following code
```

```python
null = []
for index, row in Data.iterrows():
    if row['tenure'] == '' or row['tenure'] == 'NaN':
        null.append(index)

Data = Data.drop(null)
Data
```
